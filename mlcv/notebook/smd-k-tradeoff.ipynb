```python
import matplotlib.pyplot as plt
import numpy as np

FONTSIZE = 28
FONTSIZE_SMALL = 20
COLORMAP="viridis"
COLORS = [
    "#A10035",
    "#3FA796",
    "#FEC220",
    "#0072B2",
    "#D55E00",
    "#CC79A7",
    "#56B4E9",
    "#009E73",
    "#F0E442",
    "#D3D3D3", # light gray
]
```

```python
method = "tae"

k_list = [200, 300, 400, 500, 600, 700]
thp_list = [34.38, 26.56, 35.16, 33.59, 23.44, 33.20]
energy_list = [9.20, 42.68, 90.36, 179.78, 266.44, 351.17]
```

```python
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111)

ax.plot(thp_list, energy_list, color=COLORS[0], linewidth=3)
ax.scatter(thp_list, energy_list, color=COLORS[0])

ax.tick_params(
	left = True,
	right = False ,
	labelleft = True , 
	labelbottom = True,
	bottom = True
) 
ax.set_xlabel('THP', fontsize=FONTSIZE)
ax.set_ylabel('Energy', fontsize=FONTSIZE)
ax.set_xticks([])
ax.set_yticks([])
ax.grid(linestyle="dotted", alpha=0.7)

plt.tight_layout()
plot_path = f"../img/{method}_k_tradeoff.png"
plt.savefig(plot_path, dpi=300)
plt.savefig(plot_path.replace('.png', '.pdf'))
```
